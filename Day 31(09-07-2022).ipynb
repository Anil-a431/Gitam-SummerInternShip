### Data Visualization with Pandas and Matplotlib

```python
# Histograms Charts
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
data = np.random.normal(0,20,1000)
#fixed size bin
bins = np.arange(-100,100,5)
plt.xlim([min(data)-5,max(data)+5])
plt.hist(data,bins=bins,alpha=0.5)
plt.title('Example Histogram Chart')
plt.xlabel('Bin size = 5')
plt.ylabel('Count')
plt.show()
```

```python
import math
```

```python
# Fixed Number of bins in Histogram Plot/Chart
data = np.random.normal(0,20,1000)
bins = np.linspace(math.ceil(min(data)),math.floor(max(data)),20)
plt.xlim([min(data)-5,max(data)+5])
plt.hist(data,bins=bins,alpha=0.5)
plt.title('Fixed Numbe of bins in Hist')
plt.xlabel('Variable x')
plt.ylabel('Count')
plt.show()
```

```python
# Histogram of 2 overlapping data sets
data1 = np.random.normal(0,20,1000)
data2 = np.random.normal(-100,100,1000)
plt.figure(figsize=(10,8))
bins = np.linspace(math.ceil(min(data2)),math.floor(max(data2)),20)
plt.xlim([min(data1+data2)-5,max(data1+data2)+5])
plt.hist(data1,bins=bins,alpha=0.6,label='Plot 1')
plt.hist(data2,bins=bins,alpha=0.8,label='Plot 2')
plt.title('Overlapping of 2 Histogram charts')
plt.xlabel('Varable x')
plt.ylabel('Count')
plt.legend(loc='upper right')
plt.show()
```

```python
# Scatter Plot
data1 = np.array([0,0])
data2 = np.array([[2,0],[0,2]])
x1_samples = np.random.multivariate_normal(data1,data2,100)
x2_samples = np.random.multivariate_normal(data1+0.2,data2+0.2,100)
x3_samples = np.random.multivariate_normal(data1+0.4,data2+0.4,100)

plt.figure(figsize=(10,8))
plt.scatter(x1_samples[:,0],x1_samples[:,1],marker='x',color='blue',
           alpha=0.8,label='x1 Samples')
plt.scatter(x2_samples[:,0],x2_samples[:,1],marker='o',color='red',
           alpha=0.8,label='x2 Samples')
plt.scatter(x3_samples[:,0],x3_samples[:,1],marker='p',color='green',
           alpha=0.8,label='x3 Samples')

plt.title('Scatter Example')
plt.xlabel('Variable x')
plt.ylabel('Variable y')
plt.legend(loc='upper right')
plt.show()
```

```python
# Scatter Plot with Labels
x_cords = [0.13,0.22,0.39,0.59,0.68,0.74,0.93]
y_cords = [0.75,0.34,0.44,0.52,0.80,0.25,0.55]
plt.figure(figsize=(8,6))
plt.scatter(x_cords,y_cords,marker='s',s=50)

for x,y in zip(x_cords,y_cords):
    plt.annotate(
    '(%s %s)'%(x,y),
    xy=(x,y),
    xytext=(0,-10),
    textcoords = 'offset points',
    ha='center',
    va='top')

plt.xlim([0,1])
plt.ylim([0,1])
plt.show()
```

```python
### Increasing point size with distance from the origin
fig = plt.figure(figsize=(8,6))
data1 = np.array([0,0])
data2 = np.array([[1,0],[0,1]])
x = np.random.multivariate_normal(data1,data2,500)
r = x ** 2
r_sum = r.sum(axis=1)
plt.scatter(x[:,0],x[:,1],color='red',
           marker='o',
           s=32.*r_sum,
           edgecolor='black',
           alpha=0.8)
plt.show()
```
