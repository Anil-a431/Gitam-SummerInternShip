### Sorting and Ranking in Pandas

```python
import pandas as pd
import numpy as np
```

```python
obj_series = pd.Series(range(5),
                       index=['e','b','d','c','a'])
```

```python
obj_series
```

```python
obj_series.sort_index()
```

```python
obj_series.sort_values()
```

```python
obj_series1 = pd.Series([19,1,25,11,9],
                       index=['e','b','d','c','a'])
```

```python
obj_series1.sort_values()
```

```python
df = pd.DataFrame(np.arange(12).reshape(3,4),
                 index = ['one','two','three'],
                 columns=list('dabc'))
```

```python
df
```

```python
df.sort_index()
```

```python
df.sort_index(axis=1)
```

```python
df.sort_values(by='b')
```

```python
df.sort_values(by=['b','a'])
```

```python
df.sort_values(by=['b','a'],ascending=False)
```

```python
df.sort_values(by = ['b','a'],ascending=False)
```

```python
df.sort_index(axis=1,ascending=False)
```

```python
df.sort_index(axis=1,ascending=True)
```
