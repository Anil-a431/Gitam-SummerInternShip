### Data Visualization with Pandas and Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Bar plot
# Input Values
mean_values = [1,2,3]
variance = [0.2,0.4,0.5]
bar_labels = ['bar 1','bar 2','bar 3']
```

```python
# Plot bars
x_pos = list(range(len(bar_labels)))
plt.bar(x_pos,mean_values,yerr=variance,align='center',alpha=0.5)
plt.grid()
```

```python
# Set height of y-axis
max_y = max(zip(mean_values,variance))
print(max_y)
```

```python
plt.ylim([0,(max_y[0]+max_y[1])*1.1])
```

```python
# Set the axis
plt.ylabel('Varience y')
plt.xticks(x_pos,bar_labels)
plt.title('Example of bar chart')
plt.show()
```

```python
# Bar plot
# Input Values
mean_values = [1,2,3]
variance = [0.2,0.4,0.5]
bar_labels = ['bar 1','bar 2','bar 3']
# Plot bars
x_pos = list(range(len(bar_labels)))
plt.bar(x_pos,mean_values,yerr=variance,align='center',alpha=0.5)
plt.grid()
# Set height of y-axis
max_y = max(zip(mean_values,variance))
plt.ylim([0,(max_y[0]+max_y[1])*1.1])
# Set the axis
plt.ylabel('Varience y')
plt.xticks(x_pos,bar_labels)
plt.title('Example of bar chart')
plt.show()
```

```python
mean_values = [1,2,3]
std_dev = [0.2,0.4,0.5]
bar_labels = ['bar 1','bar 2','bar 3']
fig =  plt.figure(figsize=(10,8))

# plot bars
y_pos = np.arange(len(mean_values))
y_pos = [x for x in y_pos]
plt.yticks(y_pos,bar_labels,fontsize=10)

# plot
plt.barh(y_pos,mean_values,xerr=std_dev,align='center',alpha=0.5,color='b')
plt.xlabel('Mean x')
t = plt.title('Bar plot with Horizantal')
plt.grid()
plt.show()
```

```python
# back to back bars
x1 = np.array([1,2,3,4,5,7])
x2 = np.array([2,2,3,3,3,4])
bar_labels = ['b1','b2','b3','b4','b5','b6']
fig =  plt.figure(figsize=(10,8))

y_pos = np.arange(len(x1))
plt.yticks(y_pos,bar_labels,fontsize=15)

plt.barh(y_pos,x1,align='center',alpha=0.5,color='b')

plt.barh(y_pos,-x2,align='center',alpha=0.5,color='r')

plt.xlabel('Measurement x')
t = plt.title("Back to Back Bar Plots")
plt.grid()
plt.show()
```

```python
# Line Plots
x = [1,2,3]
y_1 = [50,60,70]
y_2 = [20,30,40]

plt.plot(x,y_1,marker='*')
plt.plot(x,y_2,marker='^')

plt.xlim([0,len(x)+1])
plt.ylim([0,max(y_1+y_2)+10])

plt.xlabel('x - Axis')
plt.ylabel('Y - Axis')
plt.legend(['Sample1','Sample2'],loc='upper left')
plt.title('Line Plot Example')
plt.show()
```
