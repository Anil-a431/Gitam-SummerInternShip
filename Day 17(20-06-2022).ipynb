### Regular Expression
- Used to match the specific pattern
- Let's you want to match the pattern with Mobile Number, EmailID, Student ID and etc.
- To start with any regular expression, the first character will be cap symbol
- To end with any regular expression, the last character should be Dollar symbol
- To Match a single digits will use [0-9]
- If there is a repetation in the input, then we can use curly braces and repeat the number

```python
import re
```

```python
# Function to validate only two digit number
def validate_two_digits(n):
    pattern = '^[0-9]{2}$'
    n = str(n)
    if re.match(pattern,n):
        return True
    return False
```

```python
validate_two_digits(12)
```

```python
validate_two_digits(123)
```

```python
def validate_between_twolimits(n):
    pattern = '^[0-9]{5,15}$'
    n = str(n)
    if re.match(pattern,n):
        return True
    return False
```

```python
validate_between_twolimits(128129811219)
```

```python
validate_between_twolimits(1121)
```

```python
validate_between_twolimits(129182091802801980298019820)
```
